# 3) Figure out why the data is missing

Look at the columns **"Street Number Suffix"** and **"Zipcode"** from the [San Francisco Building Permits dataset](https://www.kaggle.com/aparnashastry/building-permit-applications-data). Both of these contain missing values.

- Which, if either, are missing because they weren't recorded?  

Once you have an answer, run the code cell below.

```python
percent_missing_suffix = sf_permits["Street Number Suffix"].isnull().sum()
percent_missing_zipcode = sf_permits["Zipcode"].isnull().sum()

# Determine the nature of the missing values
suffix_missing_reason = "Doesn't exist" if percent_missing_suffix > 0 else "Not recorded"
zipcode_missing_reason = "Doesn't exist" if percent_missing_zipcode > 0 else "Not recorded"

print("Street Number Suffix missing values reason:", suffix_missing_reason)
print("Zipcode missing values reason:", zipcode_missing_reason)

# Check your answer
q3.check()
```
